```python
import sys
import os
import pandas as pd
sys.path.append(os.path.abspath(".."))

from src.data.data_manager import DataManager
from src.data.features_engineering import unique_values, missing_values, duplicates_values
```

```python
data = DataManager("C:/Users/danil/Desktop/PROJECTS/formula1/datasets/raw/")
```

### Unique Count from results

```python
results = data.get("results")

unique_df = pd.DataFrame({
    'Category': ['Constructors', 'Drivers', 'Races'],
    'Unique Count': [
        unique_values(results['constructorId']),
        unique_values(results['driverId']),
        unique_values(results['raceId']),
    ]
})
display(unique_df)
```

### Check missing values from results

```python
categories = ['Constructors', 'Drivers', 'Races']
columns = ['constructorId', 'driverId', 'raceId']
tables = ['constructors', 'drivers', 'races']

missing_df = pd.DataFrame([
    {
        'Category': cat,
        'Missing Values': missing_values(
            results[col],
            data.get(table)[col]
        )
    }
    for cat, col, table in zip(categories, columns, tables)
])
display(missing_df)
```

### Check duplicates from results

```python
duplicates_df = pd.DataFrame([
    {
        'Category': cat,
        'Duplicate Count': duplicates_values(data.get(table)[col])
    }
    for cat, col, table in zip(categories, columns, tables)
])
display(duplicates_df)
```
